```python
#load data from JSON File
from google.colab import drive
import json


drive.mount('/content/drive')

# Path to your JSON file in Google Drive
file_path = r"/content/drive/My Drive/corpus-engineering.json"

# Load data from JSON file
with open(file_path, "r", encoding="utf-8") as file:
    data = json.load(file)


# Display the first entry to understand the structure
print(json.dumps(data[0], indent=2, ensure_ascii=False))
```

```python
from hazm import Normalizer

normalizer = Normalizer()

# Normalize each entry in the JSON data
for entry in data:
    entry['title'] = normalizer.normalize(entry['title'])
    entry['abstract'] = normalizer.normalize(entry['abstract'])
```

```python
# Check the type and structure of `data`
print(type(data))
print(data[:2])  # Print the first two entries to verify
```

```python
# Prepare documents for indexing
documents = [
    {"content": entry['abstract'], "meta": {"title": entry['title'], "FirstSubject": entry['FirstSubject'], "SecondSubject": entry['SecondSubject']}}
    for entry in data
]
```

```python
from haystack.document_stores import InMemoryDocumentStore
from haystack.nodes import BM25Retriever, FARMReader
from haystack.pipelines import ExtractiveQAPipeline

# Initialize the document store with BM25 enabled
document_store = InMemoryDocumentStore(use_bm25=True)

# Write documents to the document store
document_store.write_documents(documents)

# Initialize retriever and reader
retriever = BM25Retriever(document_store=document_store)
reader = FARMReader(model_name_or_path="HooshvareLab/bert-fa-base-uncased", use_gpu=False)

# Initialize QA pipeline
pipeline = ExtractiveQAPipeline(reader, retriever)


```

```python
# Function to query the system and print results
def query_system(query):
    prediction = pipeline.run(query=query, params={"Retriever": {"top_k": 10}, "Reader": {"top_k": 5}})
    print(f"Query: {query}\n")
    for idx, answer in enumerate(prediction['answers']):
        print(f"Answer {idx + 1}:")
        print(f"Text: {answer.answer}")
        print(f"Score: {answer.score}")
        print(f"Context: {answer.context}\n")
    return prediction

# Example query
query = "عوامل بیماری زای قارچی گیاهان دارویی"
response = query_system(query)
```
